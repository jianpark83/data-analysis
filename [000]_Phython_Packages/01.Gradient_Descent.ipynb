# II. Gradient() 정의

```python
def Gradient(x,y,w,b):
    y_hat = Machine(w,x,b)

    dw = np.mean((y-y_hat)*(-2*x))
    db = np.mean((y-y_hat)*(-2))
    return dw, db
```

```python
y = np.array([2,4,6,8,10])
dw, db = Gradient(x,y,w,b)
```

```python
print(dw)
print(db)
```

```python
def Learning(x,y,w,b,step):
    dw, db = Gradient(x,y,w,b)

    uw = w - step * dw
    ub = b - step * db
    return uw, ub
```

```python
step = 0.05
uw, ub = Learning(x,y,w,b,step)
```

```python
print(uw)
print(ub)
```

```python
def Gradient(x, y, w, b):
    y_hat = Machine(x, w, b)

    dw = np.mean((y - y_hat) * (-2 * x))
    db = np.mean((y - y_hat) * (-2))
    MSE = np.mean((y - y_hat)**2)

    return dw, db, MSE
```

```python
def Learning(x, y, w, b, step):
    dw, db, MSE = Gradient(x, y, w, b)

    uw = w - step * dw
    ub = b - step * db

    return uw, ub, MSE
```

```python
w = 2
b = 3
step = 0.005  #중요(0.001 기본)
mse = []
```

```python
for i in range(0, 3000):
    uw, ub, MSE = Learning(DATA.inputs, DATA.outputs, w, b, step)

    w = uw
    b = ub
    mse.append(MSE)
```
